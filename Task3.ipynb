```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("vgsales.csv")
df.head()
```

```python
df.sort_values("Global_Sales", ascending=False).head(10)
```

```python
plt.figure(figsize=(10,6))
sns.countplot(y='Genre', data=df)
plt.show()
```

```python
df['Platform'].value_counts()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,6))
sns.countplot(y='Platform', data=df, order=df['Platform'].value_counts().index)
plt.title("Number of Games by Platform")
plt.show()
```

```python
df[['NA_Sales','EU_Sales','JP_Sales','Other_Sales']].sum()
```

```python
region_sales = df[['NA_Sales','EU_Sales','JP_Sales','Other_Sales']].sum()

region_sales.plot(kind='bar')
plt.title("Total Video Game Sales by Region")
plt.ylabel("Sales (Millions)")
plt.show()
```

```python
df['Publisher'].value_counts().head(10)
```

```python
top_publishers = df['Publisher'].value_counts().head(10)

top_publishers.plot(kind='bar')
plt.title("Top 10 Publishers by Number of Games")
plt.ylabel("Number of Games")
plt.show()
```

```python
df.groupby('Publisher')['Global_Sales'].sum().sort_values(ascending=False).head(10)
```

```python
plt.figure(figsize=(10,6))

sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')

plt.title("Correlation Between Sales Variables")
plt.show()
```
